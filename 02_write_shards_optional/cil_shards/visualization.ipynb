## Plot distribution before and after pruning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
mypath = '../../../data/commaai/training_files/unfiltered_smaller_images/'
```

```python
df_after = pd.read_csv('../../../data/commaai/training_files/unrestricted_gauss_dens_resampled/yaws.csv', header = None)
```

```python
fig1 = plt.figure(1)
ax = fig1.add_subplot(1,1,1)
ax.hist(df_after[0], bins=100, density=False, label='Histogram from samples',
        zorder=5, edgecolor='k', alpha=0.5)
ax.set_ylim(0,20000)
plt.xlabel('steering angle in degrees')
plt.ylabel('frequency')
ax.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.savefig('angles_unrestricted_gaussian_resampled.pdf', bbox_inches = 'tight')
```

```python
df_before = pd.read_csv('../../../data/commaai/training_files_filtered/indices/review.csv')
```

```python
fig1 = plt.figure(1)
ax = fig1.add_subplot(1,1,1)
ax.hist(df_before['angle'], bins=120, density=False, label='Histogram from samples',
        zorder=5, edgecolor='k', alpha=0.5)
ax.set_ylim(0,80000)
plt.xlabel('steering angle in degrees')
plt.ylabel('frequency')
ax.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.savefig('all_angles.pdf', bbox_inches = 'tight')
```
