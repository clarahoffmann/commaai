```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x,y = np.meshgrid(np.linspace(-4,10,12),np.linspace(-4,10,12))

u = x/(1+ np.sqrt(x**2 + y**2))
v = y/(1+ np.sqrt(x**4 + y**2))

plt.quiver(x,y,u,v, color = 'grey', alpha = 0.2)
plt.vlines(-2, -4, 11, color = 'black')
plt.hlines(-2, -4,11, color = 'black')
#plt.arrow(2,0, 0,2, color = 'black')
plt.arrow(2, -2, 0, 3.95, head_width=0.15 , length_includes_head=True, color = 'grey', width = 0.01)
plt.arrow(2, 2, 1, 1.15, head_width=0.15 , length_includes_head=True, color = 'navy')
plt.arrow(3, 3.15, 1, 0.8, head_width=0.15 , length_includes_head=True, color = 'navy')
plt.arrow(4, 3.95, 1, 0.5, head_width=0.15 , length_includes_head=True, color = 'navy')
plt.arrow(6, 4.75, 0, -6.6, head_width=0.15 , length_includes_head=True, color = 'grey', width = 0.01)
plt.arrow(5, 4.45, 1, 0.35, head_width=0.15 , length_includes_head=True, color = 'navy')
plt.plot([2], [-2], marker='o', markersize=5, color="steelblue")
plt.plot([6], [-2], marker='o', markersize=5, color="steelblue")
plt.ylim(-4,11)
plt.xlim(-4,11)
plt.text(10.5, -1.5, r'$\vartheta$', fontsize=12)
plt.text(-3, 9, 'r', fontsize=12)
plt.text(1.9, 5.5, r'M leapfrog steps', fontsize=12, color = 'navy')
plt.text(6.5, 0, 'project back to\nparameter\nspace', fontsize=12, color = 'grey')
plt.text(0, -1, 'lift to\nphase\nspace', fontsize=12, color = 'grey')
plt.text(2, -3.5, 'last state', fontsize=12, color = 'steelblue')
plt.text(6, -3.5, 'new proposal', fontsize=12, color = 'steelblue')
plt.text(-1, 9, 'vector field \ngiven by Hamiltonian', fontsize=12, color = 'grey', alpha = 0.7)
#plt.text(2.5, 6, r'M leapfrog steps', fontsize=12)
plt.axis('off')
#plt.show()
plt.savefig('illustration_hmc.png', dpi = 600)
```
